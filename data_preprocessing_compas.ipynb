### Data Loading

```python
import numpy as np
import pandas as pd

df = pd.read_csv('compas-scores.csv')
df
```

```python
df = df[df['is_recid']!=-1]
df = df[df['decile_score']!=-1]
df
```

### Create Adjacency Matrix

```python
df['c_jail_in']= pd.to_datetime(df['c_jail_in'])
df['c_jail_out']= pd.to_datetime(df['c_jail_out'])
```

```python
df = df.dropna(axis=0,subset=['c_jail_in'])
df = df.dropna(axis=0,subset=['c_jail_out'])
df = df.reset_index()
df = df.drop(columns=['index'])
df
```

```python
from collections import namedtuple

Range = namedtuple('Range', ['start', 'end'])
N = df.shape[0]
matrix = np.zeros((N,N))

for i in range(N):
    r1 = Range(start=df.iloc[i,16],end=df.iloc[i,17])
    if i % 10 == 0:
        print(i)
    for j in range(i+1,N):
        r2 = Range(start=df.iloc[j,16],end=df.iloc[j,17])
        latest_start = max(r1.start, r2.start)
        earliest_end = min(r1.end, r2.end)
        delta = (earliest_end - latest_start).days
        if delta >=0:
            matrix[i,j] = 1
            matrix[j,i] = 1

```

```python
np.save("adj_matrix_2",matrix)
```
